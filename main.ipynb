```python
import os
import ast
import matplotlib.pyplot as plt
from analyzer import CodeAnalyzer

# üî¢ Complexity score calculator
def calculate_complexity_score(functions, loops, ifs, depth, is_recursive):
    score = 0
    score += functions * 2      # 2 points per function
    score += loops * 3          # 3 points per loop
    score += ifs * 2            # 2 points per if/else
    score += depth * 1          # 1 point per nesting level
    if is_recursive:
        score += 5              # 5 bonus for recursion
    return score
```

```python
# Folder containing your Python files
folder_path = "test_files"

# Store results
file_names = []
scores = []
difficulties = []

# Loop through all files
for filename in os.listdir(folder_path):
    if filename.endswith(".py"):
        filepath = os.path.join(folder_path, filename)
        with open(filepath, "r", encoding="utf-8") as f:
            source = f.read()

        # Parse and analyze
        tree = ast.parse(source)
        analyzer = CodeAnalyzer()
        analyzer.visit(tree)

        # Score calculation
        is_recursive = bool(analyzer.recursive_functions)
        score = calculate_complexity_score(
            analyzer.functions,
            analyzer.loops,
            analyzer.ifs,
            analyzer.max_depth,
            is_recursive
        )

        # Difficulty level
        if score <= 10:
            level = "üü¢ Easy"
        elif score <= 20:
            level = "üü° Medium"
        else:
            level = "üî¥ Hard"

        # Print summary
        print(f"\nüì¶ {filename}")
        print("Functions:", analyzer.functions)
        print("Loops:", analyzer.loops)
        print("Ifs:", analyzer.ifs)
        print("Recursive:", ", ".join(analyzer.recursive_functions) if is_recursive else "None")
        print("Nesting Depth:", analyzer.max_depth)
        print("üî¢ Complexity Score:", score)
        print("üéØ Difficulty Level:", level)

        # Save for chart
        file_names.append(filename)
        scores.append(score)
        difficulties.append(level)
```

```python
# Plot file scores
plt.figure(figsize=(10, 5))
bars = plt.bar(file_names, scores, color='salmon', edgecolor='black')
plt.xlabel("File")
plt.ylabel("Complexity Score")
plt.title("üìä Code Complexity Score by File")

# Annotate with difficulty labels
for bar, label in zip(bars, difficulties):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, label, ha='center', va='bottom')

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import streamlit as st
import ast
from analyzer import CodeAnalyzer

st.title("üß† Code Complexity Analyzer")

# ‚úÖ THIS LINE defines uploaded_files
uploaded_files = st.file_uploader("Upload Python files", type="py", accept_multiple_files=True)

report_data = []

for file in uploaded_files:
    source_code = file.read().decode("utf-8")
    tree = ast.parse(source_code)
    
    analyzer = CodeAnalyzer()
    analyzer.visit(tree)

    is_recursive = bool(analyzer.recursive_functions)

    score = (
        analyzer.functions * 2 +
        analyzer.loops * 3 +
        analyzer.ifs * 2 +
        analyzer.max_depth +
        (5 if is_recursive else 0)
    )

    report_data.append({
        "Filename": file.name,  # ‚úÖ Fix is here!
        "Functions": analyzer.functions,
        "Loops": analyzer.loops,
        "Ifs": analyzer.ifs,
        "Recursion": "Yes" if is_recursive else "No",
        "Nesting Depth": analyzer.max_depth,
        "Score": score,
        "Difficulty": "Easy" if score <= 10 else "Medium" if score <= 20 else "Hard"
    })
```

```python
import csv
with open("code_scores.csv", "w", newline="",encoding="utf-8") as csvfile:
    fieldnames = ["Filename", "Functions", "Loops", "Ifs", "Recursion", "Nesting Depth", "Score", "Difficulty"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for data in report_data:
        writer.writerow(data)
    print("‚úÖ Results exported to code_scores.csv")
```
